```python
import re
import pandas as pd
import numpy as np

from pathlib import Path
from datetime import datetime
```

```python
data_folder = Path('..', 'data')

data_files = list(data_folder.glob("*"))

data_files
```

```python
pattern = re.compile("(\d{9})(\d{1})(\d{2})(\d{2})(\d{4})(\d{2})(\d{2})(\w{0,1})(\d{12})([\D]+)")
```

```python
filename = data_files[0]
lines = None
count_lines = 0
count_register = 0

data = list()

with open(filename, 'r', encoding='ISO8859-1') as file:
    for line in file:
        count_lines += 1
        match = pattern.fullmatch(line)
        if match:
            count_register += 1
            data.append(match.groups())
        else:
            print(line)
        
    
print(count_lines, count_register)

columns = [
    'sequential_number',
    'cod_op',
    'day',
    'month',
    'year',
    'hour',
    'minutes',
    'cod_input',
    'pis',
    'name'
]

frame = pd.DataFrame(data, columns=columns)

del data

# frame['name'] = frame['name'].str.replace('\ {0,}\n{1,}', '')
frame['name'] = frame.name.str.strip('[ \n]')

numeric_attribs = ['cod_op', 'day', 'month', 'year', 'hour', 'minutes' ]

frame[numeric_attribs] = frame[numeric_attribs].astype(int)

date_attribs = [ 'year', 'month', 'day', 'hour', 'minutes']

frame['datetime']  = pd.to_datetime(frame[date_attribs])
```

```python
def hour_index(index):

    def registro(col):
        v = None
        try:
            v = col.iloc[index]
        except IndexRange as e :
            v = np.nan
        finally:
            return v
    
    return registro

def horas_trabalho(column):
    
    hours = column.values
    total = 0

    if len(hours) <= 6 and (len(hours) % 2 == 0):
        
        for i, f in zip(hours[0::2], hours[1::2]):
            total += (f - i)
        
        total = total
    
    return total

def registros_corretos(column):
    hours = column.values
    return len(hours) >= 0 and len(hours) <= 6 and (len(hours) % 2 == 0)

table = df.groupby(['pis', df.datetime.dt.date])['datetime'].agg([
            horas_trabalho,
            registros_corretos,
            hour_index(0), hour_index(1), hour_index(2), 
            hour_index(3), hour_index(4), hour_index(5), 
            hour_index(6), hour_index(7), 
        ]).dropna(axis=1, how='all')

table['horas_trabalho'] = pd.to_timedelta(table['horas_trabalho'])
```
